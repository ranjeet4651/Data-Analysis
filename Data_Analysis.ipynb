Joins and merges

```python
emp_dict = {
    'emp_id' : [101,102,103,104,105],
    'name' : ['Ranjeet','Basavaraj','Girija','Samanvi','Sunita'],
    'dept_id' : [1,2,1,3,None]
}
emp_df = pd.DataFrame(emp_dict)
emp_df
```

```python
dept_dict = {
    'dept_id' : [1,2,4],
    'dept_description' : ['HR','IT','Sales']
}
dept_df = pd.DataFrame(dept_dict)
dept_df
```

Inner join

```python
pd.merge(emp_df, dept_df, on='dept_id', how='inner')
```

```python
pd.merge(emp_df,dept_df, on = 'dept_id', how = 'left')
```

```python
pd.merge(emp_df, dept_df, on = 'dept_id', how = 'right')
```

```python
pd.merge(emp_df, dept_df, on = 'dept_id', how = 'outer')
```

cross join

```python
pd.merge(emp_df, dept_df, how = 'cross')
```

Grouping and Aggregation

```python
# average mpg of the cars group by cyl

cars_df.groupby('cyl')['mpg'].mean()
```

```python
cars_df.drop(columns=['Model']).groupby('cyl').mean()
```

```python
cars_df.drop(columns=['Model']).groupby('cyl').agg(['mean','max','min'])
```

```python
cars_df.drop(columns=['Model']).groupby('cyl').agg(['min','mean','max'])[['mpg','hp']]
```

```python
cars_df.drop(columns=['Model']).groupby(['cyl', 'gear']).agg(['min','mean','max'])[['mpg','hp']]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars_df['qsec'] = cars_df['qsec'].fillna(cars_df['qsec'].mean())
```

```python
cars_df.isna().sum()
```

```python
cars_df.head()
```

```python
cars_df.drop(columns=['S.No'], inplace = True)
```

```python
cars_df.head()
```

```python
cars_df.rename(columns={'Unnamed: 1' : 'Model','wt' : 'Weight'},inplace = True)
```

```python
cars_df.head()
```

```python
plt.figure(figsize=(5,3))
plt.plot(cars_df['hp'])
plt.plot(cars_df['disp'])
plt.show()
```

```python
# add a legend

plt.figure(figsize=(5,3))
plt.plot(cars_df['hp'], label = 'HP')
plt.plot(cars_df['disp'], label='Disp')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(cars_df['hp'], label='HP')
plt.plot(cars_df['disp'], label = 'Disp')
plt.legend(loc='center') # give any where you want -- 'best', 'center', 'right top','right bottom',.....
plt.show()
```

Stack plot

```python
plt.figure(figsize=(5,3))
plt.stackplot(cars_df['hp'], cars_df['disp'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
x = cars_df['disp'].index
y = cars_df['disp']
plt.stackplot(x,y)
plt.show()
```

```python
x = cars_df['disp'].index
y1 = cars_df['disp']
y2 = cars_df['hp']
plt.stackplot(x,y1, labels = 'disp')
plt.stackplot(x,y2, labels = 'hp')
plt.legend()
plt.show()
```

```python
x = cars_df['disp'].index
y1 = cars_df['disp']
y2 = cars_df['hp']
plt.stackplot(x,y1, labels = 'disp', alpha=0.7, color='#08D6C8')
plt.stackplot(x,y2, labels = 'hp', alpha=0.7, color='tomato')
plt.legend()
plt.show()
```

Bar Chart

```python
plt.figure(figsize=(5,3))
x = cars_df['cyl'].index
y = cars_df['cyl']
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,3))
x = cars_df['Model']
y = cars_df['cyl']
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,3))
x = cars_df['Model']
y = cars_df['cyl']
plt.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(4,8))
x = cars_df['Model']
y = cars_df['cyl']
plt.barh(x,y)
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(5,3))
x = cars_df['qsec']
y = cars_df['disp']
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,3))

x = cars_df['qsec']
y = cars_df['disp']

plt.scatter(x,y, marker='*', s = 50, color = 'red')

plt.xlabel('\nTime taken for quater mile')
plt.ylabel('Displacement\n')
plt.title('Qsec vs Disp\n',fontsize=16, fontweight='bold')
plt.show()
```

# Subplots

```python
plt.figure(figsize=(5,3))

plt.scatter(cars_df.qsec , cars_df.disp, marker='*', s = 50, color = 'red')
plt.stackplot(cars_df.disp.index, cars_df.disp, labels='hp', alpha=0.7, color='#CC6CE7')
```

```python
plt.figure(figsize=(8,3))

#subplot
plt.subplot(1,2,1)  # 1 row, 2 columns, 1st subplot
plt.scatter(cars_df.qsec , cars_df.disp, marker='*', s = 50, color = 'red')

plt.subplot(1,2,2) # 1 row, 2 columns, 2nd subplot
plt.stackplot(cars_df.disp.index, cars_df.disp, labels='hp', alpha=0.7, color='#CC6CE7')
```

```python
plt.figure(figsize=(5,6))

#subplot
plt.subplot(2,1,1)  # 2 row, 1 columns, 1st subplot
plt.scatter(cars_df.qsec , cars_df.disp, marker='*', s = 50, color = 'red')

plt.subplot(2,1,2) # 2 row, 1 columns, 2nd subplot
plt.stackplot(cars_df.disp.index, cars_df.disp, labels='hp', alpha=0.7, color='#CC6CE7')
```

# **Pie chart**

```python
animals = ['dogs','cat','wolf','lion']
weights = [30,20,50,90]

plt.pie(weights, labels = animals , autopct='%1.1f%%') # autopct -- value 1.1 after point how many decimals
plt.show()
```

```python
animals = ['dogs','cat','wolf','lion']
weights = [30,10,50,90]

plt.pie(weights, labels = animals , autopct='%1.0f%%') # autopct -- value 1.1 after point how many decimals
plt.title('Pie chart')
plt.show()
```

```python
import seaborn as sns
```
